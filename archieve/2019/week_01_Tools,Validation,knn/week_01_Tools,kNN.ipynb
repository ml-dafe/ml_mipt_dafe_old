**Matplotlib** — Great library for visualizations.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 10, 20)
```

```python
fig = plt.figure(figsize=(10, 6))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, x ** 2, 'r')
axes.plot(x, x ** 3, 'b*--')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
axes.legend([r'$x^2$', 'x^3'], loc = 0)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (16, 5))

for pow_num, ax in enumerate(axes):
    ax.plot(x, x ** (pow_num + 1), 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(r'$y = x^' + str(pow_num + 1)+ r'$', fontsize = 18)
fig.tight_layout() # dealing with intersecting figures
```

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)

phi_m = np.linspace(0, 2 * np.pi, 100)
phi_p = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(14, 6))

ax = fig.add_subplot(111, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0, cmap='jet')
```

# kNN - метод ближайших соседей

```python
from sklearn.datasets import load_iris
```

```python
X = pd.DataFrame(load_iris()['data'], columns=load_iris()['feature_names'])
X.head()
```

```python
print(load_iris()['target_names'])
y = load_iris()['target']
y
```

```python
plt.scatter(X['sepal length (cm)'], X['petal length (cm)'], c=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[['sepal length (cm)', 'petal length (cm)']], y,
                                                    train_size=0.8, test_size=0.2,
                                                    shuffle=True, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
knn = KNeighborsClassifier() # По умочанию - 5 соседей и Евклидова метрика
```

```python
knn.fit(X_train, y_train)
```

```python
pred = pd.DataFrame(knn.predict(X_test), columns=['predict'])
pred['test'] = y_test
```

```python
pred
```

```python
accuracy_score(knn.predict(X_test), y_test)
```

```python
knn = KNeighborsClassifier(1)
knn.fit(X_train, y_train)
accuracy_score(knn.predict(X_test), y_test)
```

```python
knn = KNeighborsClassifier(80)
knn.fit(X_train, y_train)
accuracy_score(knn.predict(X_test), y_test)
```
